```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
model=load_model('model.h5')

with open('one_hot_encoder_geography.pkl','rb') as file:
    one_hot_encoder_geography=pickle.load(file)
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)
with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geo_encoded=one_hot_encoder_geography.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df
```

```python
input_df=pd.concat([input_df.drop('Geography',axis=1), geo_encoded_df],axis=1)
```

```python
input_df
```

```python
input_scaled=scaler.transform(input_df)
```

```python
input_scaled
```

```python
prediction=model.predict(input_scaled)
prediction_prob=prediction[0][0]
prediction_prob
```

```python
if prediction_prob>0.5:
    print('This customer is likely to churn')
else:
    print('This customer is not likely to churn')
```
